```python
import gym
```

```python
env = gym.make('gym_go:go-v0', size=4, komi=0)
```

```python
observation = env.reset()
for i in range(100):
    action = env.uniform_random_action()
    state, reward, done, info = env.step(action)
    env.render('terminal')
    if env.game_ended():
        print(f'Num moves: {i}')
        break
```
